```python
profile_nlist = []
rating_vlist=[]
profile_rlist=[]

for url in df['Company_link']:
    response = requests.get(url)


    soup = BeautifulSoup(response.text, 'html.parser')
    reviews = soup.find_all('div', class_='review')
    

    for review in reviews[:20]:
        
        profile_name = review.find('div', class_='b')
        try:
            profile_name = profile_name.find('span', itemprop='name').text.strip()
            profile_name_list.append(profile_name)
            
            rating_div = review.find('div', class_='rate rate_5', itemprop='reviewRating')
           
            rating_value = float(rating_div.find('span', itemprop='ratingValue').get_text())
            rating_value_list.append(rating_value)
        
            profile_reviews= review.find('div',class_='text').get_text()
            profile_reviews_list.append(profile_reviews)

        except:
            profile_name=profile_nlist.append("NO Name Found")
            rating_value= rating_vlist.append('No Rating found')
            profile_reviews=profile_rlist.append("No Reviews Data")

review_dict = {'Profile Name': profile_nlist[:140],
                       'Review Text': profile_rlist,
                       'Rating': rating_vlist}


# dump the review data to a JSON file
with open('reviews.json', 'w') as f:
    json.dump(review_dict, f, indent=2)
```
